```python
import io
import boto3
import tifffile
from botocore.config import Config
from botocore import UNSIGNED

s3 = boto3.client("s3", config=Config(signature_version=UNSIGNED))


def download_tiff(s3_key, bucket="cellpainting-gallery"):
    """
    Given an S3 key to a TIFF file, return it as a NumPy array.
    """
    response = s3.get_object(Bucket=bucket, Key=s3_key)
    tif_data = response["Body"].read()
    img = tifffile.imread(io.BytesIO(tif_data))
    return img


import numpy as np


def download_npy(s3_key, bucket="cellpainting-gallery"):
    """
    Given an S3 key to an NPY file, return it as a NumPy array.
    """
    response = s3.get_object(Bucket=bucket, Key=s3_key)
    npy_data = response["Body"].read()
    arr = np.load(io.BytesIO(npy_data))
    return arr


def apply_illum_correction(raw_img, illum_function, method="division"):
    """
    Apply illumination correction to a raw image using the given illumination function.

    method = "division" or "subtraction"
    """
    if method == "division":
        # To avoid dividing by zero, add a small epsilon where illum_function is zero
        epsilon = 1e-8
        corrected = raw_img / (illum_function + epsilon)
    elif method == "subtraction":
        corrected = raw_img - illum_function
        corrected[corrected < 0] = 0  # clip negative values
    else:
        raise ValueError(
            "Unknown correction method. Choose 'division' or 'subtraction'."
        )
    return corrected
```

```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.transform import resize as sk_resize


def show_raw_vs_corrected(
    df,
    row_index,
    channel_short_name,  # e.g., "DNA", "ER", "RNA", "Mito", "AGP", etc.
    bucket="cellpainting-gallery",
    method="division",
    int_percentile=99.9,
    resize=None,
):
    """
    Download a raw channel image + its illumination function from a given row
    and display them side-by-side (raw vs. corrected).

    Args:
        df (pd.DataFrame):
            DataFrame that has columns "Image_<channel_short_name>" and "Illum_<channel_short_name>".
        row_index (int):
            Which row index in df to visualize.
        channel_short_name (str):
            e.g. "DNA", "ER", "RNA", "Mito", "AGP", "Brightfield", etc.
        bucket (str):
            S3 bucket name. Defaults to "cellpainting-gallery".
        method (str):
            Illumination correction method, "division" (typical for fluorescent channels) or "subtraction" (often for brightfield).
        int_percentile (float):
            Percentile for intensity clipping in display. E.g., 99.9 means clip at the 99.9th percentile.
        resize (tuple or None):
            If a tuple (height, width), images will be resized to that shape for display.
            If None (default), no resizing is performed.
    """

    row_data = df.iloc[row_index]

    image_col = f"Image_{channel_short_name}"
    illum_col = f"Illum_{channel_short_name}"

    if image_col not in df.columns:
        raise ValueError(f"Column {image_col} not found in DataFrame.")
    if illum_col not in df.columns:
        raise ValueError(f"Column {illum_col} not found in DataFrame.")

    raw_s3_key = row_data[image_col]
    illum_s3_key = row_data[illum_col]

    if not isinstance(raw_s3_key, str) or pd.isna(raw_s3_key):
        raise ValueError(f"No valid path for {image_col} at row {row_index}.")
    if not isinstance(illum_s3_key, str) or pd.isna(illum_s3_key):
        raise ValueError(f"No valid path for {illum_col} at row {row_index}.")

    print(f"Row index: {row_index}")
    meta_cols = ["Metadata_Plate", "Metadata_Well", "Metadata_Field"]
    meta_cols = [c for c in meta_cols if c in df.columns]
    if meta_cols:
        print(row_data[meta_cols].dropna().to_string())

    raw_img = download_tiff(raw_s3_key, bucket=bucket)
    illum_function = download_npy(illum_s3_key, bucket=bucket)
    corrected_img = apply_illum_correction(raw_img, illum_function, method=method)

    if resize is not None:
        # resize expects (rows, cols) i.e. (height, width)
        raw_img = sk_resize(
            raw_img, resize, preserve_range=True, anti_aliasing=True
        ).astype(raw_img.dtype)
        corrected_img = sk_resize(
            corrected_img, resize, preserve_range=True, anti_aliasing=True
        ).astype(corrected_img.dtype)

    # For display, let's scale each image from 0 to int_percentile
    def scale_for_display(img):
        vmax = np.percentile(img, int_percentile)
        vmax = max(vmax, 1e-6)
        return np.clip(img / vmax, 0, 1)

    raw_scaled = scale_for_display(raw_img)
    corrected_scaled = scale_for_display(corrected_img)

    # Show side by side
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    ax_raw, ax_corrected = axes

    ax_raw.imshow(raw_scaled, cmap="gray")
    ax_raw.set_title(f"Raw {channel_short_name}")
    ax_raw.axis("off")

    ax_corrected.imshow(corrected_scaled, cmap="gray")
    ax_corrected.set_title(f"Corrected {channel_short_name} ({method})")
    ax_corrected.axis("off")

    plt.tight_layout()
    plt.show()
```

```python
for channel in ["DNA", "ER", "RNA", "Mito", "AGP"]:
    show_raw_vs_corrected(df, row_index=200, channel_short_name=channel)
```

We can clearly see that the illumination corrections are working by looking at the corners of the images. In the raw images, corners are darker than the middle of the image, but in the corrected image we can see that the corners are brighter!
